```python
test_per_compl = df[df['test preparation course']== 'completed']['test preparation course'].count()
test_per_compl
```

```python
## how many students are NOT completed there test preparation
```

```python
test_per_not_compl = df[df['test preparation course']== 'none']['test preparation course'].count()
test_per_not_compl
```

```python
334+665          #count
```

```python
t_p_c = test_per_compl*100/999      #percentage of test preparation completed
t_p_c
```

```python
t_p_n_c = test_per_not_compl*100/999    #percentage of test preparation NOT completed
t_p_n_c
```

```python
33.433433433433436+66.56656656656656      
```

```python
plt.pie([t_p_c,t_p_n_c], labels=['test preparation completed','test preparation NOT completed'], shadow= True)
```
